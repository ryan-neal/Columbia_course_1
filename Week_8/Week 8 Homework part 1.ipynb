```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import datetime
import seaborn as sns
%matplotlib inline
```

```python
def fix_nan(data):
    try:
        data = str(data)
        return datetime.datetime.strptime(data,'%m/%d/%Y %I:%M:%S %p')
    except:
        return np.NaN
```

```python
df = pd.read_csv("311_data.csv")
```

```python
df.head()
```

```python
df['Created Date'] = df['Created Date'].apply(lambda x:datetime.datetime.strptime(x,'%m-%d-%Y %I:%M:%S'))
df["Closed Date"] = df["Closed Date"].apply(fix_nan)
df["processing_time"] = df["Closed Date"] - df["Created Date"]
df["start_time_window"] = df["Created Date"].apply(lambda x:x.hour)
lst = ["Closed Date", "Created Date", "Agency", "Borough", "processing_time", "start_time_window"]
df[lst].to_csv("output1.csv" )
```
