```python
file = open("boogiewoogie.html", "r").read()
```

```python
results_page = BeautifulSoup(file,'lxml')
print(results_page.prettify())
```

```python
musicians = []
for ul in results_page.find_all('ul'):
    for li in ul.find_all("li"):
        for a in li.find_all("a"):
            musicians.append("https://en.wikipedia.org" + a.get("href"))
            break
zz_spot = [i[0] for i in enumerate(musicians) if "ZZ" in i[1]]
musicians = musicians[22: zz_spot[0] + 1]
print(musicians)
```

```python
def read_ps(filename):
    file2 = open(filename, "r")
    results = BeautifulSoup(file2, "lxml")
    all_text = ""
    for tag in results.find_all('p'):
        all_text += tag.get_text() 
    return(all_text)
```

```python
def get_all_text(directory):
    import os
    all_text = []
    dictionary = {}
    for file in os.listdir(directory):
        if file[0] != ".":
            dictionary[file.split(".")[0]] = read_ps(directory + "/" + file)
    return dictionary
```

```python
def conversion(dictionary):
    newdict = {}
    counter = 0
    for key in dictionary.keys():
        newdict[counter] = key
        counter += 1
    return newdict
```

```python
all_text = get_all_text("samples")
```

```python
test_text = get_all_text("test")
```

```python
def create_model(dictionary):
    from gensim import corpora
    from gensim.models.ldamodel import LdaModel
    from gensim.parsing.preprocessing import STOPWORDS
    texts = [[word for word in text.lower().split()
        if word not in STOPWORDS and word.isalnum()]
        for text in dictionary.values()]
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    from gensim.similarities.docsim import Similarity
    from gensim import models, similarities
    lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
    return lsi
```

```python
create_model(all_text)
```

```python
import pandas as pd
lst = []
for key, value in test_text.items():
    doc = value
    vec_bow = dictionary.doc2bow(doc.lower().split())
    vec_lsi = lsi[vec_bow]
    index = similarities.MatrixSimilarity(lsi[corpus])
    sims = index[vec_lsi]
    sims = sorted(enumerate(sims), key=lambda item: -item[1])
    x = converter[sims[0][0]]
    y = key
    lst.append([y,x])
    
```

```python
answer = pd.DataFrame(lst, columns = ["TestName", "SampleName"])
```

```python
def run():
    import pandas as pd
    def get_all_text(directory):
        import os
        all_text = []
        dictionary = {}
        for file in os.listdir(directory):
            if file[0] != ".":
                dictionary[file.split(".")[0]] = read_ps(directory + "/" + file)
        return dictionary

    def conversion(dictionary):
        newdict = {}
        counter = 0
        for key in dictionary.keys():
            newdict[counter] = key
            counter += 1
        return newdict
    
    all_text = get_all_text("samples")
    test_text = get_all_text("test")
    
    def create_model(dictionary):
        from gensim import corpora
        from gensim.models.ldamodel import LdaModel
        from gensim.parsing.preprocessing import STOPWORDS
        texts = [[word for word in text.lower().split()
            if word not in STOPWORDS and word.isalnum()]
            for text in dictionary.values()]
        dictionary = corpora.Dictionary(texts)
        corpus = [dictionary.doc2bow(text) for text in texts]
        from gensim.similarities.docsim import Similarity
        from gensim import models, similarities
        lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
        return lsi
    
    create_model(all_text)
    
    lst = []
    for key, value in test_text.items():
        doc = value
        vec_bow = dictionary.doc2bow(doc.lower().split())
        vec_lsi = lsi[vec_bow]
        index = similarities.MatrixSimilarity(lsi[corpus])
        sims = index[vec_lsi]
        sims = sorted(enumerate(sims), key=lambda item: -item[1])
        x = converter[sims[0][0]]
        y = key
        lst.append([y,x])
        
    answer = pd.DataFrame(lst, columns = ["TestName", "SampleName"])
    answer.to_csv("output.csv", index=False)
```

```python
run()
```
